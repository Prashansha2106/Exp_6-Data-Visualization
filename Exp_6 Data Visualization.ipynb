```python
# importing the basic library
import numpy as np
from matplotlib import pyplot as plt
```

#line Chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

#Bar chart

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# generate random numbers
data = np.random.randint(1, 20, 10)

# bar chart with color
plt.bar(range(len(data)), data, color='black')   # single color
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# generate random numbers
data = np.random.randint(1, 20, 10)

# list of colors for each bar
colors = ['red', 'blue', 'green', 'purple', 'orange', 
          'cyan', 'pink', 'yellow', 'gray', 'brown']

# bar chart with multiple colors
plt.bar(range(len(data)), data, color=colors)

plt.title("Random Numbers Bar Chart", fontsize=14)
plt.xlabel("Index")
plt.ylabel("Value")

plt.show()
```
